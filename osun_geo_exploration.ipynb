```python
import pandas as pd
import numpy as np
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon
df = pd.read_csv('only osun.csv')
df.head()
```

```python
nga = gpd.read_file(r'new_lga_nigeria_2003.shp')
```

```python
osun.to_crs(epsg=4326).plot()
```

```python
#reformatting our data into a GeoPandas Dataframe

crs = {'init':'EPSG:4326'}
geometry = [Point(xy) for xy in zip(df['longitude'], df['latitude'])]
geo_df = gpd.GeoDataFrame(df, 
                          crs = crs, 
                          geometry = geometry)
```

```python
geo_df.head()
```

```python
geo_df.columns
```

```python
fig, ax = plt.subplots(figsize = (10,10))
osun.to_crs(epsg=4326).plot(ax=ax, color='lightgrey')
geo_df.plot(ax=ax)
ax.set_title('All listed points')
```

```python
fig, ax = plt.subplots(figsize = (10,10))
osun.to_crs(epsg=4326).plot(ax=ax, color='lightgrey', edgecolor='black', column='lga')
geo_df.plot(column = 'status_id', ax=ax, cmap = 'rainbow',
            legend = True, 
            markersize = 10)
ax.set_title('Osun Water Status ID HeatMap')
```

```python
geo_df.status_id.isnull().sum()
```

```python
fig, ax = plt.subplots(figsize = (10,10))
osun.to_crs(epsg=4326).plot(ax=ax, color='lightgrey', edgecolor='black', column='lga')
geo_df.plot(column = 'water_source', ax=ax, cmap = 'rainbow',
            legend = True, 
            markersize = 10)
ax.set_title('Osun Water Source HeatMap')
```

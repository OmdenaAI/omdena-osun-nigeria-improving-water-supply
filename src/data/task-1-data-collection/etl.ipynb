#### Proposal
That to every water point there should be selected four resident locations (located at the boundaries of the water location). This will ease the problem of finding how far water points are from residents. 

Resident's Longitude and Latitude together with the water point's location longitude and latitude can be used to calculate the distance between the resident and the water points.

```python

def distance(lat1, lat2, lon1, lon2):
     
    # The numpy module contains a function named
    # radians which converts from degrees to radians.
    lon1a = np.radians(lon1)
    lon2a = np.radians(lon2)
    lat1a = np.radians(lat1)
    lat2a = np.radians(lat2)
      
    # Haversine formula
    
    dlon = np.radians(lon2 - lon1)
    dlat = np.radians(lat2 - lat1)
    a = np.sin(dlat / 2)**2 + np.cos(lat1a) * np.cos(lat2a) * np.sin(dlon / 2)**2
 
    #c = 2 * asin(sqrt(a))
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    r = 6371
      
    # calculate the result
    return np.round(c * r, 6)
```

```python
distance_in_km = []
for row in df.itertuples(index=False):
   distance_in_km.append(
       distance(row.resident_latitude, row.latitude, row.resident_longitude, row.longitude)
   )
```

```python
df['distance_in_km'] = distance_in_km
```
